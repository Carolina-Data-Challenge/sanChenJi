```python
import pandas as pd
```

```python
nc_2018 = pd.read_csv(r"C:\Users\Wayne\OneDrive - University of North Carolina at Chapel Hill\CDC 2020\house_purchases.csv")
```

```python
nc_2018['county_code'] = nc_2018['county_code'].fillna(0)
```

```python
nc_loans = nc_2018[(nc_2018['county_code'] >= 37001) & (nc_2018['county_code'] <= 37199)]
```

```python
nc_loans[["county_code"]] = nc_loans[["county_code"]].astype(int)
```

```python
county_names = pd.read_csv(r"C:\Users\Wayne\Downloads\fips.csv",usecols = ['County','FIPS code'], low_memory = False)
```

```python
county_names['FIPS code'] = county_names['FIPS code'] + 37000
```

```python
county_names['County'] = county_names['County'].str.replace(' County', '')
```

```python
merged = pd.merge(nc_loans, county_names,how='left',left_on = 'county_code', right_on = 'FIPS code')
```
